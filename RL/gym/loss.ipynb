```python
# 画一个-log(x)函数
```

```python
import numpy as np

import matplotlib.pyplot as plt

# 功能描述：绘制-log(x)函数的曲线
# 输入：无（直接在区间(0, 2]上绘制）
# 输出：显示函数图像

x = np.linspace(0.01, 2, 100)  # 避免log(0)导致无穷
y = -np.log(x)

plt.plot(x, y)
plt.title('-log(x)函数图像')
plt.xlabel('x')
plt.ylabel('-log(x)')
plt.grid(True)
plt.show()
```
