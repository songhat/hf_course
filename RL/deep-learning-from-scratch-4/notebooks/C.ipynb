```python
samples = 10000
action_size = 4

Qs = []

for _ in range(samples):
    Q = np.random.randn(action_size)
    Qs.append(Q.max()) 


# 在有误差的情况下,每次选最大值导致期望偏差.这种情况叫过估计.
# 但是其实际上我们的值都是在从正态分布中采样的,尽管取了最大值也应该具备随机性才对
plt.hist(Qs, bins=16)
plt.axvline(x=0, color='red')
plt.axvline(x=np.array(Qs).mean(), color='cyan') 
plt.show()
```

```python
samples = 10000
action_size = 4

Qs = []

for _ in range(samples):
    Q = np.random.randn(action_size)
    Q_prime = np.random.randn(action_size)
    idx = np.argmax(Q)
    Qs.append(Q_prime[idx])

plt.hist(Qs, bins=16)
plt.axvline(x=0, color='red')
plt.axvline(x=np.array(Qs).mean(), color='cyan')
plt.show()
```
